```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
data = np.array([48, 50, 52, 54, 49, 51, 57, 58, 55, 120, 130])
```

```python
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

```

```python
IQR = Q3-Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 - 1.5*IQR

outliers = data[(data<lower_bound) | (data>upper_bound)]
```

```python
print(outliers)
```

```python
plt.figure()
box = plt.boxplot(data)
plt.show()
```

```python
z_scores = zscore(data)
outliers = data[np.abs(z_scores)>2]
print(outliers)
```

```python
import numpy as np
# np.random.seed(861) #for reproducibility
X = np.random.randint(1, 100000, size=5)
print(X)
```

```python
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
X_scaled = s.fit_transform(X.reshape(-1,1))
print(X_scaled)
```

##MEAN

```python
s,n= 0, 0
for i in X:
    s+= i
    n=n+1
mean = s/len(X)
print(mean)
```

```python
smean = np.mean(X)
print(smean)
```

##Standard deviation

```python
sq = 0
for i in X:
    sq = sq + ((i-mean)**2)
std = (sq/len(X))**(1/2)
print(std)
```

```python
sdeviation = np.std(X)
print(sdeviation)
```

```python
X_scaled = []
for i in X:
    X_scaled.append((i-mean)/std)
print(X_scaled)
```
